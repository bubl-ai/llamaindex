## Evaluate

```python
eval_data = rag.eval_data["lyft_10k"]["train"]
eval_data.head(2)
```

### Using user defined functions

```python
from bubls.utils.evaluation.from_dataframe import is_hit, reciprocal_ranking
eval_data["is_hit"] =  eval_data.apply(is_hit, axis=1)
eval_data["reciprocal_ranking"] =  eval_data.apply(reciprocal_ranking, axis=1)
eval_data.head(2)
```

### Using LlamaIndex core evaluators

```python
from llama_index.core.evaluation import (
    FaithfulnessEvaluator,
    RelevancyEvaluator,
    AnswerRelevancyEvaluator,
    ContextRelevancyEvaluator,
    GuidelineEvaluator,
    CorrectnessEvaluator,
    SemanticSimilarityEvaluator,
    BatchEvalRunner,
)
from deepeval.integrations.llama_index import (
    DeepEvalAnswerRelevancyEvaluator,
    DeepEvalFaithfulnessEvaluator,
    DeepEvalContextualRelevancyEvaluator,
    DeepEvalSummarizationEvaluator,
    DeepEvalBiasEvaluator,
    DeepEvalToxicityEvaluator,
)

judges = {
    "faithfulness": FaithfulnessEvaluator(),  # Measure if the response from a query engine matches any source nodes.
    "relevancy": RelevancyEvaluator(),  # Measure if the response + source nodes match the query.
    "answer_relevancy": AnswerRelevancyEvaluator(),  # measure on the relevancy of a generated answer to a given user query. (Does response match and tries to address the subject matter?) 
    "context_relevancy": ContextRelevancyEvaluator(),  # measure on the relevancy of retrieved contexts to a given user query. (Does response match and tries to address the subject matter?)
    # "deepeval_summarization": DeepEvalSummarizationEvaluator(),  # DeepEval presents an opinionated framework for unit testing RAG applications. It breaks down evaluations into test cases, and offers a range of evaluation metrics that you can freely evaluate.
    "guideline": GuidelineEvaluator(guidelines = "The response must be respectful."),  # evaluate a question answer system given user specified guidelines.
    "correctness": CorrectnessEvaluator(),  # relevance and correctness of a generated answer against a reference answer
    "semantic_similarity": SemanticSimilarityEvaluator(),  # calculates the similarity score between embeddings of the generated answer and the reference answer.
}

from bubls.utils.evaluation.from_dataframe import evaluators_df
eval_data[list(judges.keys())] = eval_data.apply(evaluators_df, axis=1, args=(judges,))
eval_data.head(2)
```

### Using LlamaIndex core evaluators, running in batches

```python
runner = BatchEvalRunner(judges, workers=4)
eval_results = await runner.aevaluate_queries(
    rag.query_engine["lyft_10k"],
    queries=eval_data["query"].values.tolist(),
    reference=eval_data["reference"].values.tolist(),
)
```
